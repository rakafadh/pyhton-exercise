**PLatform mana yang rata-rata daily minutes spend nya paling banyak**

```python
df1 = pd.read_csv('E:/Main Room/ITworkerwannabe/Data scientist wannabe/datasets/social_media_usage.csv', index_col='App')

instagram = df1.loc[['Instagram'], ['Daily_Minutes_Spent']]
facebook = df1.loc[['Facebook'], ['Daily_Minutes_Spent']]
twitter = df1.loc[['Twitter'], ['Daily_Minutes_Spent']]
snapchat = df1.loc[['Snapchat'], ['Daily_Minutes_Spent']]
pinterest = df1.loc[['Pinterest'], ['Daily_Minutes_Spent']]
linkedin = df1.loc[['LinkedIn'], ['Daily_Minutes_Spent']]

mean_instagram = instagram.mean()
mean_facebook = facebook.mean()
mean_twitter = twitter.mean()
mean_snapchat = snapchat.mean()
mean_pinterest = pinterest.mean()
mean_linkedin = linkedin.mean()

platform_mean = [mean_instagram, mean_facebook, mean_twitter, mean_snapchat, mean_pinterest, mean_linkedin]
platform_name = ['Instagram', 'Facebook', 'Twitter', 'Snapchat', 'Pinterest', 'LinkedIn']

platform_mean = [round(mean[0], 2) for mean in platform_mean]

plt.figure(figsize=(10, 10))
plt.pie(platform_mean, labels=platform_name, autopct='%1.1f%%')

```
