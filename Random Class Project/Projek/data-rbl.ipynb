```python
'''Regresi data Hasil Percobaan Mobil Bertenaga Pendulum Dengan Variasi Massa Bandul'''
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np


x = [45,45,45,60,60,60,75,75,75]
y = [23,64,79,42,77,91,52,69,92]

plt.figure(figsize = (10,3))
plt.scatter(x, y, color="green")
plt.xlabel("Sudut Simpangan(derajat)")
plt.ylabel("Panjang Lintasan(cm)")
plt.title("Regresi Hubungan Sudut Simpangan dengan Panjang Lintasan")
plt.grid()
plt.show()


```
