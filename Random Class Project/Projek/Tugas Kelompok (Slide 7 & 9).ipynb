```python
# Halaman   : 9
# Soal      : 1

import pandas as pd
import matplotlib.pyplot as plt
dffin = pd.read_csv("E:/ITB KEOS/intro to computation/financial.csv")

#Menghitung Nilai Rata-Rata 
df1 = dffin.loc[(dffin["Country"] == "United States of America") & (dffin["ManufacturingPrice"])].mean()["ManufacturingPrice"]
df2 = dffin.loc[(dffin["Country"] == "United States of America") & (dffin["SalePrice"])].mean()["SalePrice"]

#Mengelompokkan Data sebelum diplot
x = [df1,df2]
y = ["Manufacturing \nPrice","Sale \nPrice"]

#proses membuat visualisasi
plt.figure(figsize=(3,7))
plt.bar(y, x, color='lightcoral')
plt.title('Perbandingan Rata-Rata \nManufacturing & Sale Price \nUnited States of America', size=16)
plt.ylabel('nilai', size=14)
plt.xticks(size=12)
plt.yticks(size=12)
plt.show()

#Penjelasan pemilihan Bar Chart
print("note:\nJenis grafik Bar Chart dipilih karena Jika anda ingin membandingkan nilai antar beberapa kategori, \nmaka grafik yang paling ideal adalah bar chart sebab dapat dengan mudah menunjukkan \nnilai terkecil dan terbesar dari satu set data")
```
