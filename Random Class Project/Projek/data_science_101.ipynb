```python
#Menghitung rata rata pendapatan dari peminjam

data2 = pd.read_csv("loan_sanction_train.csv", index_col="CoapplicantIncome")
mean = data2.mean()["ApplicantIncome"]

print(f"Rata-Rata Pendapatan dari Peminjam adalah {mean}")
```

```python
#Menentukan gender yang paling banyak melakukan pinjaman

#Menghitung jumlah pinjaman peminjam perempuan
data3 = pd.read_csv("loan_sanction_train.csv", index_col="Gender")
female = data3.loc[["Female"], ["LoanAmount"]]
sum_female = female.sum()["LoanAmount"]
print(f"Jumlah pinjaman peminjam bergender perempuan yakni {sum_female}")
print("sedangkan,")

#Menghitung jumlah pinjaman peminjam laki-laki
male2 = data3.loc[["Male"], ["LoanAmount"]]
sum_male = male2.sum()["LoanAmount"]
print(f"Jumlah pinjaman peminjam bergender laki-laki yakni {sum_male}")

#Menyimpulkan
print("Berdasarkan data tersebut,\ndisimpulkan bahwa gender yang paling banyak mengajukan pinjaman adalah : ")
print(f"Gender Laki-Laki dengan pinjaman total sejumlah {sum_male}")
```

```python
#Menghitung rata-rata pinjaman yang diajukan untuk masing-masing area properti
#Rata-rata area Urban
data4 = pd.read_csv("loan_sanction_train.csv", index_col="Property_Area")
urban = data4.loc[["Urban"], ["LoanAmount"]]
mean_urban = urban.mean()["LoanAmount"]
print(f"Rata-rata area Urban adalah {mean_urban}")

#Rata-rata area SemiUrban
data4 = pd.read_csv("loan_sanction_train.csv", index_col="Property_Area")
semiurban = data4.loc[["Semiurban"], ["LoanAmount"]]
mean_semiurban = semiurban.mean()["LoanAmount"]
print(f"Rata-rata area SemiUrban adalah {mean_semiurban}")

#Rata-rata area Rural
data4 = pd.read_csv("loan_sanction_train.csv", index_col="Property_Area")
rural = data4.loc[["Rural"], ["LoanAmount"]]
mean_rural = rural.mean()["LoanAmount"]
print(f"Rata-rata area Rural adalah {mean_rural}")
```

```python
import pandas as pd
import  matplotlib.pyplot as plt
#Histogram Penghasilan Peminjam, dengan judul, dan label untuk tiap sumbu yang jelas!
data5 = pd.read_csv("loan_sanction_train.csv")
x = data5["ApplicantIncome"]

plt.figure(figsize= (10,3))
plt.hist(x, color= "green")
plt.ylabel("Jumlah Orang")
plt.xlabel("Penghasilan")
plt.title("Histogram Penghasilan Peminjam")
plt.grid()
```

```python
data5 = pd.read_csv("loan_sanction_train.csv")
y = len(data5.loc[data5["Education"] == "Graduate"])
z = len(data5.loc[data5["Education"] == "Not Graduate"])

education = [y, z]
edu_label = ["Sudah Lulus","Tidak/Belum Lulus"]
explodes = [0.1,0]

plt.figure(figsize=(6,6))
plt.axes()
plt.pie(education, labels= edu_label, autopct='%1.0f%%', explode=explodes)
plt.title("Tingkat Pendidikan Peminjam")
```
